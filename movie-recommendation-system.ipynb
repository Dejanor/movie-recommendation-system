```python
import ast
def convert (obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
data['genres']= data['genres'].apply(convert)
data['keywords']= data['keywords'].apply(convert)
```

```python
def convert3(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
        else:
            break
    return l
```

```python
data['cast']= data['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            l.append(i['name'])
            break
    return l
```

```python
data['crew']= data['crew'].apply(fetch_director)
```

```python
 data['overview']= data['overview'].apply(lambda x:x.split())
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer ()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.stem import PorterStemmer

# Assuming `data` is your DataFrame and `new_df` is defined properly

# Initialize the vectorizer and the stemmer
tfidf_vectorizer = TfidfVectorizer()
ps = PorterStemmer()

# Stemming function
def stem(text):
    return ' '.join([ps.stem(word) for word in text.split()])

# Combine text fields
data['combined_text'] = data['title'] + ' ' + data['tag'].apply(lambda x: ' '.join(x))

# Fit the TF-IDF vectorizer
tag_matrix = tfidf_vectorizer.fit_transform(data['combined_text'])
feature_matrix = tag_matrix

# Get user input and process it
user_tag = input("Enter movie name or anything related to the movie: ").lower()
user_tag = stem(user_tag)
user_vector = tfidf_vectorizer.transform([user_tag])
user_feature = user_vector

# Calculate cosine similarities
similarities = cosine_similarity(user_feature, feature_matrix)

# Get top recommendations
top_indices = similarities.argsort(axis=1)[:, -10:][0][::-1]
top_recommendations = data.iloc[top_indices]['title']

# Print recommendations
print("Top Recommendations: ")
for i, title in enumerate(top_recommendations):
    print(f"{i+1}. {title}")
```
